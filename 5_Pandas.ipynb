## 並べ替え/sorting

* データフレームのある列を抽出し, `df.sort_values()` メソッドを呼び出すことで値の**並べ替え (sort)** を行うことができます.

* なお, このメソッドは並べ替えが終わったあとの値で, もとのデータフレーム内の値を置き換えることまでは行わず, 結果を返します.

* そこで, 別の変数で結果を受け取り, 始めの 5 行を表示することで並べ替えが行われたことを確認してみます.

* `df.sort_values()` は, デフォルトでは**昇順 (ascending)** に並べ替えを行います.

* 昇順とは, だんだん値が大きくなっていくように並べ替えるときの並べ方のことで, 逆にだんだん値が小さくなっていくように並べ替えるときは、**降順 (descending)** に並べると言います.

* `df.sort_values()` は並べ替えを行いたい列の名前を `by` という引数で受け取ります.

* また, デフォルトでは昇順に並べ替えを行います.

* You can extract a column of a data frame and call the method `df.sort_values()` to **sort** the values.

* Note that this method does not replace the values in the original data frame by the values after sorting, but returns the result.

* Check that the sorting has been done by receiving the result in another variable and displaying the first 5 rows.

* sort_values()` sorts in **ascending** order by default.

* Ascending order means that the values are sorted in increasing order, **descending order** means that the values are sorted in decreasing order.

* The `df.sort_values()` takes an argument `by` which is the name of the columns to sort by.

* Sort in ascending order by default.

```python
# total_rooms 列の値を昇順に並べ替え
df_as = df.sort_values(by='total_rooms')
```

```python
df_as.head()
```

* 降順に並べ替える場合は, `ascending=False` という引数の指定を行います.

* To sort in descending order, specify the argument `ascending=False`.

```python
# total_rooms の列の値を降順に並べ替え/Sort total_rooms column values in descending order
df_de = df.sort_values(by='total_rooms', ascending=False)
```

```python
df_de.head()
```

* 着目したい要素や, 行, 列を選択する方法を紹介します.

* 入力値 `x` と目標値 `t` が別の列として同じ配列に格納されている場合は, 指定した列だけを取り出す操作を行って, 結果を別の変数に代入する操作を行います.

* このように, 特定の列の選択や列を範囲指定して選択する機能が Pandas のデータフレームにも用意されています.

* 今回は最後の列 `median_house_value` と, それ以外の列をそれぞれ取り出して, 別々の変数に格納してみます.

* 列や行の選択を行う方法は複数あります.

* ここでは, 整数インデックスを用いてデータの部分選択を行う `df.iloc[]` を紹介します.

* Here is how to select the element, row or column you want to focus on.

* In case the input value `x` and the target value `t` are stored in the same array as different columns, you can perform an operation to take out only the specified column and assign the result to another variable.

* Thus, the selection of a specific column or a range of columns is also provided in Pandas data frames.

* In this example, we take the last column `median_house_value` and the other columns and store them in separate variables.

* There are several ways to select columns and rows.

* Here we introduce `df.iloc[]` which performs partial data selection using integer indexes.

```python
# データの確認/Data Verification
df.head(3)
```

```python
# df.iloc[行, 列]/df.iloc[row, column]
# 0 行目 longitude 列の選択/Select row 0 longitude column
df.iloc[0, 0]
```

```python
# 1 行目 latitude 列の選択/Select row 1, column latitude
df.iloc[1, 1]
```

* `iloc` には, NumPy の ndarray の中の値を部分的に選択するのと同様のスライス表記を用いることができます.

* 負のインデックスを使い, 末尾の要素からの個数を用いて位置指定を行うこともできます.

* `iloc` can use a slice notation similar to NumPy's ndarray for partial selection of values.

* You can also use negative indices, and specify the position using the number of elements from the last element.

```python
# すべての行の、最後の列を選択/Select the last column of all rows
t = df.iloc[:, -1]
```

```python
# 先頭3件の表示/Show first 3 records
t.head(3)
```

```python
# 型の確認/Confirmation of type
type(t)
```

* If only one row or one column is extracted, a **series (series)** object is returned.

```python
# すべての行の、先頭の列から末尾の列のひとつ手前までを選択
# Select all rows from the first column to the one before the last column
x = df.iloc[:, 0:-1]
```

```python
# すべての行の、先頭の列から末尾の列のひとつ手前までを選択
# Select all rows from the first column to the one before the last column
x = df.iloc[:, :-1]
```

## 条件指定による要素の選択/Selecting Elements by Conditional Specification

* 値に対する条件を指定してデータの選択を行う方法を紹介します.

* `median_house_value` 列を選択し, 返ってきたシリーズオブジェクトに対して, 比較演算子を使い**各要素に対する条件**を指定し, 条件を満たすかどうかを全要素に対して調べた結果を取得します.

* Here is how to select data by specifying conditions on values.

* Select the `median_house_value` column, specify **conditions for each element** using the comparison operator on the returned series object, and get the result of checking all elements to see if they satisfy the conditions.

```python
# median_house_value 列を選択し、全要素に対し 70000 より大きいかどうかを計算
# Select column median_house_value and calculate if it is greater than 70000 for all elements
mask = df['median_house_value'] > 70000
```

```python
mask.head()
```

* 比較演算子の片方の辺にデータフレームやシリーズをおくと, 指定された条件を満たすかどうかを全ての要素に対して計算することができます.

* 結果は, 各要素が条件を満たすか, 満たさないかを表す `True`、`False` が各要素の位置に格納されたデータフレームやシリーズとなります.

* これを**マスク (mask)** と呼ぶことがあります.

* データフレームやシリーズも NumPy の ndarray と同様に, マスクを使って要素を選択することができます.

* `mask` を `df` に `[]` を使って与えることで, 指定した条件を満たす要素だけを取り出すことができます.

* If you put a data frame or series on one side of the comparison operator, you can compute for all elements whether they satisfy the specified condition or not.

The result is a data frame or series with `True` and `False` at the position of each element, indicating * whether the element satisfies or does not satisfy the condition.

* This is sometimes called **mask**.

* Data frames and series can be selected by using masks, just like NumPy's ndarray.

* By giving `mask` to `df` with `[]`, you can extract only the elements that satisfy the specified condition.

```python
# df[mask] の先頭 5 件を表示/Showing first 5 items of df[mask]
df[mask].head()
```

### 複数の条件指定による要素の選択/Selecting elements by specifying multiple conditions

* 複数の条件を組み合わせて要素を選択することも出来ます.

* その場合は条件式を `()` でくくって用います.

* **論理和 (or)** は `|`, **論理積 (and)** は `&` を用いて表します.

* You can also combine multiple conditions to select elements.

* In this case, use `()` to enclose the conditional expression.

* **OR (or)** is denoted by `|`, **OR (and)** by `&`.

```python
# 70000 より小さい または 80000 より大きい/Less than 70000 or greater than 80000
mask2 = (df['median_house_value'] < 70000) | (df['median_house_value'] > 80000)
```

```python
mask2.head()
```

```python
df[mask2].head()
```

```python
# 70000 より大きい かつ 80000 より小さい/Greater than 70000 and less than 80000
mask3 = (df['median_house_value'] > 70000) & (df['median_house_value'] < 80000)
```

```python
mask3.head()
```

```python
df[mask3].head()
```

* 条件に当てはまる要素を調べる操作と, 条件に当てはまる要素の選択まで, 1 行にまとめて書くこともできます.

* You can also combine the operations of checking the elements that meet the conditions and selecting the elements that meet the conditions in a single line.

```python
df[(df['median_house_value'] > 70000) & (df['median_house_value'] < 80000)].head()
```

### 条件指定による要素の置換/Replacing Elements by Conditional Specification

* 条件を指定して選択した要素に対し, 値の書き換えを行うことができます.

* 例えば, `median_house_value` 列に対していくつかの条件を別々に調べ, それぞれの条件を満たしている場合に特定の値を持つような新しい列を `df` に追加します.

* `median_house_value` が下記の値を持つ `target` という列を追加します.

  - 60000 より小さい場合は 0
  - 60000 以上 70000未満は 1
  - 70000 以上 80000未満は 2
  - 80000 以上は 3

* 何も値の入っていない`target` という列を`df` に追加します.

* You can rewrite the value of a selected element by specifying a condition.

* For example, you can check several conditions on a `median_house_value` column separately, and add a new column to `df` that has a specific value if each condition is satisfied.

* Add a column called `target` whose `median_house_value` has the following values.

  - 0 if the value is less than 60000
  - 1 if between 60000 and 70000
  - 2 for greater than 70000 but less than 80000
  - 3 for values greater than 80000

* Add a column called `target` to `df` with no value in it.

```python
# 新しい列 target を None で初期化/Initialize new column target with None
df['target'] = None
```

```python
df.head()
```

* `target` という列の全ての要素は `None` となっています.

* この値を条件指定によって書き換えます.

* まず各条件に対応するマスクを作成します.

* All elements of the column `target` are `None`.

* You can rewrite this value by specifying conditions.

* First, create a mask for each condition.

```python
mask1 = df['median_house_value'] < 60000
mask2 = (df['median_house_value'] >= 60000) & (df['median_house_value'] < 70000)
mask3 = (df['median_house_value'] >= 70000) & (df['median_house_value'] < 80000)
mask4 = df['median_house_value'] >= 80000
```

* 行や列を整数インデックスで選択する場合`df.iloc[]`を使用しますが, **列を名前で指定する**場合には`df.loc[]` を用います.

* 上で計算したマスクと名前による列指定を組み合わせて, 各条件を満たす行の `target` 列の値を書き換えます.

* Use `df.iloc[]` to select rows or columns by integer index, but use `df.loc[]` to **specify columns by name**.

* Combining the mask computed above and the column specification by name, rewrite the values of the `target` columns of the rows satisfying each condition.

```python
df.loc[mask1, 'target'] = 0
df.loc[mask2, 'target'] = 1
df.loc[mask3, 'target'] = 2
df.loc[mask4, 'target'] = 3
```

* 結果を確認してみます.

* Check the results.

```python
# 先頭から 5 番目までを表示/Display from the first to the fifth
df.head()
```

## 欠損値の除去・補間/Removal and interpolation of missing values

* 欠損値を含むデータの場合, 一部の行の値が欠損している列に `NaN` (Not a Number)、`None`, `NaT` (Not a Time) などが含まれる場合があります.

* 欠損値への対策としては, 欠損値を含む行または列を除去するか, 欠損値を特定の値で補完するという方法が考えられます.

* 欠損値の除去の方法を紹介します.

* In case of data with missing values, some columns may contain `NaN` (Not a Number), `None`, `NaT` (Not a Time), etc., where some rows have missing values.

* The countermeasures against missing values are to remove rows or columns that contain missing values, or to complement the missing values with specific values.

* Here is how to remove missing values.

```python
# 欠損値を人為的に作成/Create missing values artificially
df.iloc[0, 0] = None
```

```python
# (0, 'longitude') の要素が NaN になっていることを確認/Check if (0, 'longitude') element is NaN.
df.head(3)
```

```python
# 欠損値のあるレコードを削除/Delete records with missing values
df_dropna = df.dropna()

# 先頭から 3 件を表示/Showing first 3 records
df_dropna.head(3)
```

* 上の結果と見比べると, `NaN` を含んでいた 0 行目のデータが取り除かれていることが分かります.

* 次に, 平均を使った欠損値の補完を行ってみます.

* まずは, 補完に使用する平均値の計算を行います.

* Comparing with the above result, you can see that the data in row 0, which contained `NaN`, has been removed.

* Next, we try to complete the missing values using the mean.

* First, we compute the mean value to be used for the completion.

```python
mean = df.mean()
mean
```

* 計算した各列の値の平均が格納されている `mean` を`df.fillna()` メソッドに渡すことにより, `mean` を用いた欠損値の補完を行うことができます.

* You can use `mean` to complete missing values by passing `mean`, which contains the average of the computed values of each column, to the method `df.fillna()`.

```python
# 欠損値を mean で補完/Missing values are filled with mean
df_fillna =  df.fillna(mean)

# 先頭から 3 件を表示/Showing first 3 rows
df_fillna.head(3)
```

* 0行目のデータの`longitude`列に, `mean`の`longitude`行の値が表示されていることが分かります.

* 今回は欠損値が1箇所にだけあるデータを用いましたが, `df.dropna()`や`df.fillna()`は, 対象の全ての欠損値に対して上記のような操作を行うメソッドです.

* You can see that the value of the `longitude` row of the `mean` is displayed in the `longitude` column of the data in row 0.

* In this case, we used data with only one missing value, but `df.dropna()` and `df.fillna()` are methods to do the above operation for all the missing values.
